```python
import pandas as pd
```

```python
df = pd.read_csv('accidents.csv',sep = ",")
```

```python
df
```

```python
dfsp = df.loc[df['ocorrencia_uf']=='SP']; dfsp.head()
```

```python
#agrupa por tipo de classificação
dfClassificacao = dfsp.groupby(['ocorrencia_classificacao']).count()
#remove e cria um indice novo
#dfClassificacao = dfClassificacao.reset_index()
```

```python
dfClassificacao.plot(kind = 'pie',x='ocorrencia_classificacao', y = 'codigo_ocorrencia', legend = False, title = 'Classificação de ocorrencias no estado de SP', figsize = (5,5))
```

```python
dfsp["ocorrencia_tipo"].value_counts().plot.barh(figsize=(12,12),title = 'Maiores Causas De Acidentes No Estado De SP')
```

```python
dfspmaiores = dfsp.groupby('ocorrencia_cidade').count().sort_values(by=['codigo_ocorrencia'], ascending = False).head(10)

dfspmaiores['codigo_ocorrencia'].plot(kind='bar', title = 'As 10 Cidades Com maior numero de acidentes aereos no estado de SP')
```

# Conclusão:
#### Feito a análise dos dados acima, foi analisado a quantidade de ocorrências de acidentes aéreos em território do Estado de São Paulo.
